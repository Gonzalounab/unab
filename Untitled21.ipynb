```python
import yfinance as yf
import matplotlib.pyplot as plt

# Descargar datos históricos de Tesla (TSLA)
tesla_data = yf.download('TSLA', start='2010-06-29', end='2024-11-08')

# Restablecer el índice para convertir la fecha en una columna normal
tesla_data.reset_index(inplace=True)

# Función para crear un gráfico
def make_graph(data, column, title, ylabel):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data[column], label=column, color='blue')
    plt.title(title)
    plt.xlabel('Fecha')
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid()
    plt.show()

# Gráfico 1: Precio de Cierre
make_graph(tesla_data, 'Close', 'Gráfico de Precio de Cierre de Tesla (TSLA)', 'Precio de Cierre (USD)')

# Gráfico 2: Volumen de Transacciones
make_graph(tesla_data, 'Volume', 'Gráfico del Volumen de Transacciones de Tesla (TSLA)', 'Volumen de Transacciones')
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Descargar datos históricos de GameStop (GME)
gme_data = yf.download('GME', start='2002-02-13', end='2024-11-08')

# Restablecer el índice para convertir la fecha en una columna normal
gme_data.reset_index(inplace=True)

# Función para crear un gráfico
def make_graph(data, column, title, ylabel):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data[column], label=column, color='purple')
    plt.title(title)
    plt.xlabel('Fecha')
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid()
    plt.show()

# Gráfico 1: Precio de Cierre de GameStop
make_graph(gme_data, 'Close', 'Gráfico de Precio de Cierre de GameStop (GME)', 'Precio de Cierre (USD)')

# Gráfico 2: Volumen de Transacciones de GameStop
make_graph(gme_data, 'Volume', 'Gráfico del Volumen de Transacciones de GameStop (GME)', 'Volumen de Transacciones')
```
